# Population Data

```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
p_file='./data/PEP_2018_PEPAGESEX_with_ann.csv'
```

```python
df = pd.read_csv(p_file, sep=',', error_bad_lines=False, index_col=False, encoding='latin-1',dtype='unicode')
```

```python
df_WA_traf = pd.read_csv('./data/WA_Cngstn.csv')
```

```python
df_zip = pd.read_csv('./data/zip_code_database.csv')
```

```python
df_popDen = pd.read_csv('./data/WA_popDensity.csv', header = None)
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df['Geography'][1].split(', ')[1]
```

```python
df['Geography'][1].split(' County, ')[0]
```

```python
def sep_state(x):
    return x.split(', ')[1]
def sep_county(x):
    return x.split(' County, ')[0]
```

```python
df['State'] = df['Geography'].apply(lambda x : sep_state(x))
```

```python
df['County'] = df['Geography'].apply(lambda x : sep_county(x))
```

```python
df_WA = df[df['State'] == 'Washington']
```

```python
df_WA = df_WA[['Population Estimate (as of July 1) - 2018 - Both Sexes; Median age (years)', 'Population Estimate (as of July 1) - 2018 - Male; Median age (years)', 'Population Estimate (as of July 1) - 2018 - Female; Median age (years)', 'State', 'County']]
```

```python
df_WA
```

```python
df_popDen.head()
```

```python
def get_sqmi_as_float(x):
    return x.split('/')[0]

def get_county(x):
    return x.split(',')[0]
```

```python
df_popDen['Density(/sqmi)'] = df_popDen[0].apply(lambda x : get_sqmi_as_float(x)).astype(float)
df_popDen['County'] = df_popDen[1].apply(lambda x : get_county(x))
```

```python
df_popDen.drop([0, 1], inplace=True, axis =1)
```

```python
df_popDen.head()
```

# Weather Data

```python
df_weather = pd.read_csv('./data/WA_wthr_pruned.csv')
```

```python
df_weather.head()
```

```python
df_graph = df_weather.copy()
df_graph['StartTime(UTC)'] = pd.to_datetime(df_graph['StartTime(UTC)'])
dates = df_graph['StartTime(UTC)'].dt.date
```

```python
df_2017 = df_graph[(dates>pd.Timestamp('2016-12-31').date()) & (dates<pd.Timestamp('2018-01-01'))]
df_2017.head()
```

```python
df_2017.drop(['LocationLat', 'LocationLng', 'EndTime(UTC)'], axis = 1, inplace=True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df_2017['TypeEncoded'] = le.fit_transform(df_2017['Type'])
```

```python
typ_label = sorted(df_2017.TypeEncoded.unique())
typ_legend = list(le.inverse_transform(sorted(df_2017.TypeEncoded.unique())))
print (list(zip(typ_label, typ_legend)))
```

```python
# le2 = preprocessing.LabelEncoder()
df_2017['SeveEncoded'] = le.fit_transform(df_2017['Severity'])
```

```python
sev_label = sorted(df_2017.SeveEncoded.unique())
sev_legend = list(le.inverse_transform(sorted(df_2017.SeveEncoded.unique())))
print (list(zip(sev_label, sev_legend)))
```

```python
df_2017_prune = df_2017.drop(['Type', 'Severity', 'County'], axis = 1)
```

```python
df_2017_prune['date'] = df_2017_prune['StartTime(UTC)'].dt.date
```

```python
df_2017_prune
```

```python
df_2017_prune.groupby(['ZipCode', 'date']).mean()
```

```python
df_2017_hr = df_2017_prune.copy()
df_2017_hr.set_index('StartTime(UTC)', inplace=True)
df_2017_hr = pd.DataFrame(df_2017_prune.reset_index().groupby(['ZipCode', pd.Grouper(key='StartTime(UTC)', freq='H')])['wthr_Duration', 'TypeEncoded', 'SeveEncoded'].mean())
df_2017_hr.fillna(0, inplace=True)
```

```python
df_2017_hr
```

##### full weather

```python
full_wthr = df_weather.copy()
full_wthr.drop(['LocationLat', 'LocationLng', 'EndTime(UTC)'], axis = 1, inplace=True)
full_wthr['StartTime(UTC)'] = pd.to_datetime(full_wthr['StartTime(UTC)'])

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
full_wthr['TypeEncoded'] = le.fit_transform(full_wthr['Type'])
typ_label = sorted(full_wthr.TypeEncoded.unique())
typ_legend = list(le.inverse_transform(sorted(full_wthr.TypeEncoded.unique())))
print (list(zip(typ_label, typ_legend)))

full_wthr['SeveEncoded'] = le.fit_transform(full_wthr['Severity'])
sev_label = sorted(full_wthr.SeveEncoded.unique())
sev_legend = list(le.inverse_transform(sorted(full_wthr.SeveEncoded.unique())))
print (list(zip(sev_label, sev_legend)))

full_wthr = full_wthr.drop(['Type', 'Severity', 'County'], axis = 1)
full_wthr['date'] = full_wthr['StartTime(UTC)'].dt.date
```

```python
full_wthr
```

```python
full_wthr_hr = full_wthr.copy()
full_wthr_hr.set_index('StartTime(UTC)', inplace=True)
full_wthr_hr = pd.DataFrame(full_wthr_hr.reset_index().groupby(['ZipCode', pd.Grouper(key='StartTime(UTC)', freq='H')])['wthr_Duration', 'TypeEncoded', 'SeveEncoded'].mean())
full_wthr_hr.fillna(0, inplace=True)
```

```python
full_wthr_hr
```

```python
full_wthr_hr.to_csv('./data/hourly/full_wthr_hr.csv', index=False)
```

```python
noEvent = []
totalEvent = []
for zipcode in trfc_wthr_cnss['ZipCode'].unique().tolist():
    eventPerZip = []
    for i in range (-1, 7, 1):
        e = trfc_wthr_cnss[(trfc_wthr_cnss['TypeEncoded'] == i) & (trfc_wthr_cnss['ZipCode'] == zipcode)].describe()['Duration']['mean']
        eventPerZip.append(e)
    totalEvent.append(eventPerZip)
```
